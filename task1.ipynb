```python
import pandas as pd
#importing data set
csv_file = '/content/drive/My Drive/DataScience/pima-indians-diabetes.csv'
pima_data = pd.read_csv(csv_file)
```

```python
new_column_names = [
    'NumberOfTimesPregnant',
    'PlasmaGlucose2H',
    'DiastolicBloodPressure',
    'TricepsSkinfoldThickness',
    '2HourSerumInsulin',
    'BMI',
    'DiabetesPedigreeFunction',
    'Age',
    'ClassVariable'
]
pima_data.columns = new_column_names
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'Age' is a column in your dataset
age_data = pima_data['Age']

# Histogram
sns.histplot(age_data, kde=False, color='blue', bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages in Pima Indians Diabetes Dataset')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'DiastolicBloodPressure' is a column in your dataset
diastolic_bp_data = pima_data['DiastolicBloodPressure']

# Histogram for Diastolic Blood Pressure distribution
sns.histplot(diastolic_bp_data, kde=False, color='coral', bins=20)
plt.xlabel('Diastolic Blood Pressure')
plt.ylabel('Frequency')
plt.title('Distribution of Diastolic Blood Pressure in Pima Indians Diabetes Dataset')
plt.show()
```
